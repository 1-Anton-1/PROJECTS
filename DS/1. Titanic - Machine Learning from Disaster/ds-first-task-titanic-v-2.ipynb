```python
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
titanic_data = pd.read_csv("/kaggle/input/titanic/train.csv")
titanic_data.head()
```

```python
X = titanic_data.drop(['PassengerId', 'Survived', 'Name', 'Ticket', 'Cabin'], axis=1) # Удаляем лишние колонки
y = titanic_data.Survived
X
```

```python
X = pd.get_dummies(X) # Изменяем строковую переменную на номинативную
X.head()
```

```python
X = X.fillna({'Age': X.Age.median()}) # Заменяем значения NaN в колонке Age на медианное значение
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
rf = RandomForestClassifier(n_estimators=1000, max_depth=5)
rf.fit(X_train, y_train)
```

```python
rf.score(X_test, y_test)
```

```python
test_data = pd.read_csv("/kaggle/input/titanic/test.csv")
test_data.head()
```

```python
X_test_fin = test_data.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)
X_test_fin = pd.get_dummies(X_test_fin)
X_test_fin
```

```python
X_test_fin = X_test_fin.fillna({'Age': X_test_fin.Age.median(), 'Fare': X_test_fin.Fare.median()}) # Заменяем значения NaN в колонке Age на медианное значение
X_test_fin
```

```python
predictions = rf.predict(X_test_fin)
```

```python
output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
output.to_csv('submission.csv', index=False)
print("Success!")
```
